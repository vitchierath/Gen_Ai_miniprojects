```python
import os
import gradio as gr
from duckduckgo_search import DDGS
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.schema import SystemMessage, HumanMessage

# Configure Gemini API Key
GEMINI_API_KEY = "your api"
llm = ChatGoogleGenerativeAI(model="gemini-2.0-pro", google_api_key=GEMINI_API_KEY)

# Search Function
def search_and_summarize(query):
    """Fetches search results and summarizes them using LangChain & Gemini."""
    with DDGS() as ddgs:
        results = [r['body'] for r in ddgs.text(query, max_results=5)]

    search_text = "\n".join(results)
    messages = [
        SystemMessage(content="You are an AI assistant that summarizes web content."),
        HumanMessage(content=f"Summarize this: {search_text}")
    ]
    summary = llm.invoke(messages)
    return summary.content if summary else "No summary available."

# Gradio UI
iface = gr.Interface(
    fn=search_and_summarize,
    inputs="text",
    outputs="text",
    title="🔍 AI-Powered Search Engine with LangChain & Gemini 2.0",
    description="Enter a search query and get AI-powered summaries powered by Gemini 2.0!"
)

iface.launch(share=True)
```

```python
import os
import gradio as gr
from duckduckgo_search import DDGS
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain.schema import SystemMessage, HumanMessage
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.llms import CTransformers

# Configure Gemini 2.0 API Key
GEMINI_API_KEY = "your api"
llm = ChatGoogleGenerativeAI(model="gemini-2.0-pro", google_api_key=GEMINI_API_KEY)

# # LLaMA Model Setup
# llama_llm = CTransformers(model="model/llama-2-7b-chat.ggmlv3.q8_0.bin",
#                           model_type="llama",
#                           config={"max_new_tokens":256, "temperature":0.01})



# LLaMA Model Setup
llama_llm = CTransformers(model="TheBloke/Llama-2-7B-Chat-GGML",
                          model_type="llama",
                          config={"max_new_tokens":256, "temperature":0.01})

# Blog Post Generator using LLaMA
def generate_blog_post(topic, blog_style, no_words):
    template = f"""
        Write a blog for {blog_style} job profile for a topic {topic}
        within {no_words} words.
    """
    response = llama_llm(template)
    return response if response else "No blog post available."

# AI Chatbot with Custom Context
def chatbot_response(query):
    messages = [HumanMessage(content=query)]
    response = llm.invoke(messages)
    return response.content if response else "No response available."

# Separate Gradio Interfaces
iface_blog_generator = gr.Interface(fn=generate_blog_post,
                                    inputs=["text", "text", "text"],
                                    outputs="text",
                                    title="📝 Automated Blog Post Generator with LLaMA")

iface_chatbot = gr.Interface(fn=chatbot_response,
                             inputs="text",
                             outputs="text",
                             title="🤖 AI Chatbot with Custom Context")

# Launch interfaces
iface_blog_generator.launch(share=True)
iface_chatbot.launch(share=True)

```
