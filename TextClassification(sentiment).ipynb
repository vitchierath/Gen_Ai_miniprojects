Text classification (eg: positive/negative)

```python
import gradio as gr
from transformers import pipeline

# Load the pre-trained text classification pipeline
classifier = pipeline("text-classification")

def classify_text(text):
    # Use the classifier to predict the class of the input text
    prediction = classifier(text)
    return prediction

# Define the Gradio interface
iface = gr.Interface(
    fn=classify_text,  # The function to wrap
    inputs="text",     # Input type
    outputs="json"     # Output type
)

# Launch the Gradio interface
iface.launch()
```

**Sentiment Analysis**

```python
import gradio as gr
from transformers import pipeline

# Load the pre-trained sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

def classify_text(text):
    # Use the classifier to predict the sentiment of the input text
    prediction = classifier(text)
    return prediction

# Define the Gradio interface
iface = gr.Interface(
    fn=classify_text,  # The function to wrap
    inputs=gr.Textbox(lines=5, placeholder="Enter text..."),
    outputs="json" ,
    title="Sentiment analysis",
    description="Enter a piece of text to classify its sentiment."
)

# Launch the Gradio interface
iface.launch()
```
